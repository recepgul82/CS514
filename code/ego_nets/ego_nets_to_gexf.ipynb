```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
DATA_PATH= "data/one_deg_net.txt"
GEXF_PATH = "g_data/"
```

```python
# Load the graph from the file
# source traget weight
G = nx.read_weighted_edgelist(DATA_PATH, create_using=nx.DiGraph())
print(G.number_of_nodes())

#drop the nodes with degree lower than 2
#G.remove_nodes_from(list(nx.isolates(G)))
#remove = [node for node,degree in dict(G.degree()).items() if degree <= 2]
#print(len(remove))

#G.remove_nodes_from(remove)

print(G.number_of_nodes())

```

```python
nx.write_gexf(G, GEXF_PATH + "one_deg_net.gexf")
```

```python
# plot degree distribution
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degreeCount = {}
for d in degree_sequence:
    degreeCount[d] = degreeCount.get(d, 0) + 1

deg, cnt = zip(*degreeCount.items())
plt.xlim(0, 100)
plt.xticks(range(0, 100, 10))
plt.yscale('log')
plt.bar(deg, cnt, width=0.8)
plt.title("Degree Histogram")
plt.show()
```
