```python
import csv

# Initialize an empty dictionary
names_dict = {}

# Open the CSV file and read its contents with UTF-8 encoding
with open('edit_history.csv', mode='r', encoding='utf-8') as file:
    csv_reader = csv.reader(file)
    # Extract the header
    headers = next(csv_reader)
    
    # Iterate through each row in the CSV file
    for row in csv_reader:
        # Use the first column (names) as the key in the dictionary
        name = row[0]
        # Initialize the values of each key as empty lists
        names_dict[name] = []

# Print the resulting dictionary
print(len(names_dict))
```

```python
import requests
from bs4 import BeautifulSoup

def get_politician_party_history(name):
    # Convert the politician's name to a format suitable for URL
    url_name = name.replace(" ", "_")
    url = f"https://tr.wikipedia.org/wiki/{url_name}"
    
    # Get the Wikipedia page content
    response = requests.get(url)
    
    if response.status_code != 200:
        return None
    
    # Parse the page content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the "Kişisel Bilgiler" section
    personal_info = soup.find('table', {'class': 'infobox'})

    if not personal_info:
        return None
    
    # Find the "Partisi" row within the "Kişisel Bilgiler" section
    party_row = None
    for row in personal_info.find_all('tr'):
        header = row.find('th')
        if header and 'partisi' == header.get_text().strip().lower():
            party_row = row
            break

    if not party_row:
        return None
    
    # Extract the party information
    party_cell = party_row.find('td')

    if not party_cell:
        return None
    
    # Extract party names (text within <a> tags) and exclude citations
    party_links = party_cell.find_all('a')
    party_list = [link.get_text().strip().replace(" ", "_") for link in party_links if 'cite' not in link.get('href', '')]

    return party_list

# Assuming `names_dict` is already defined and populated with politician names
# Extract party history for each politician
for name in names_dict.keys():
    party_history = get_politician_party_history(name)
    if party_history:
        names_dict[name] = party_history

# Print the resulting dictionary
for name, parties in names_dict.items():
    print(f"{name}: {parties}")
```

```python
# Create the TXT file
with open('politicians_previous_parties.txt', mode='w', encoding='utf-8') as file:
    for name, parties in names_dict.items():
        # Write the name first
        line = name
        # Add each party separated by a comma
        for party in parties:
            line += f",{party}"
        # Write the line to the file
        file.write(line + "\n")

print("TXT file 'politicians_previous_parties.txt' created successfully.")
```
